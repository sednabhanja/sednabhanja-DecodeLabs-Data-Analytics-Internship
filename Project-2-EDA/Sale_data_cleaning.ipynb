# Project 2: Exploratory Data Analysis (EDA)

## Objective

The objective of this project is to explore and analyze the dataset using descriptive statistics and visualizations to identify patterns, trends, correlations, and outliers.

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
df2 = pd.read_excel("Dataset for Data Analytics.xlsx")
```

```python
df2.head()
```

```python
df2.shape
```

```python
df2.info()
```

```python
df2.isnull().sum()
```

```python
df2.describe().T
```

```python
df2.describe(include='object').T
```

```python
df2.columns
```

```python
df2.duplicated().sum()
```

Product Distribution

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df2, x="Product")
plt.xticks(rotation=45)
plt.title("Product Distribution")
plt.show()
```

Payment Methods Distribution

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df2, x="PaymentMethod")
plt.title("Payment Methods")
plt.show()
```

Order Status Distribution

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df2, x="OrderStatus")
plt.title("Order Status")
plt.show()
```

Referral Source Distribution

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df2, x="ReferralSource")
plt.xticks(rotation=30)
plt.show()
```

QUALITY

```python
plt.figure(figsize=(7,5))
sns.histplot(df2["Quantity"], bins=20, kde=True)
plt.title("Quantity Distribution")
plt.show()
```

UNIT PRICE

```python
plt.figure(figsize=(7,5))
sns.histplot(df2["UnitPrice"], bins=20, kde=True)
plt.title("Unit Price Distribution")
plt.show()
```

TOTAL PRICE

```python
plt.figure(figsize=(7,5))
sns.histplot(df2["TotalPrice"], bins=20, kde=True)
plt.title("Total Price Distribution")
plt.show()
```

BOXPLOT FOR TOTAL PRICE

```python
plt.figure(figsize=(6,5))
sns.boxplot(y=df2["TotalPrice"])
plt.title("Outliers in Total Price")
plt.show()
```

BOXPLOT FOR UNIT PRICE

```python
plt.figure(figsize=(6,5))
sns.boxplot(y=df2["UnitPrice"])
plt.title("Outliers in Unit Price")
plt.show()
```

BOXPLOT FOR QUALITY

```python
plt.figure(figsize=(6,5))
sns.boxplot(y=df2["Quantity"])
plt.title("Outliers in Quantity")
plt.show()
```

CORREALTION ANALYSIS

```python
numeric_df = df2.select_dtypes(include=np.number)
```

HEATMAP

```python
plt.figure(figsize=(8,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

BUSINESS INSIGHTS :: Top 10 Customers by Total Spending

```python
top_customers = (
    df2.groupby("CustomerID")["TotalPrice"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

top_customers.plot(kind="bar", figsize=(10,5))
plt.title("Top 10 Customers by Total Spending")
plt.ylabel("Total Spending")
plt.show()
```

Total Sales by Product

```python
sales = df2.groupby("Product")["TotalPrice"].sum()

sales.plot(kind="bar", figsize=(10,5))
plt.title("Total Sales by Product")
plt.ylabel("Sales")
plt.show()
```

Average Total Price by Payment Method

```python
sns.barplot(data=df2,
            x="PaymentMethod",
            y="TotalPrice")
plt.title("Average Total Price by Payment Method")
plt.show()
```
